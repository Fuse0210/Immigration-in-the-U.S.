In this notebook, a bar chart that shows percentages of immigrants' home countries  will be created. 

Data was obtained from U.S. Census: 

https://data.census.gov/cedsci/table?q=u.s.%20foreign%20born%20population&tid=ACSST5Y2019.S0502&hidePreview=false

```python
import pandas as pd
import numpy as np
```

Read the dataset and modefiy it to create a bar chart

```python
df = pd.read_csv('Foreign born countries.csv')
df
```

```python
df.rename(columns = {'Unnamed: 0': 'Countries'}, inplace = True)
df
```

```python
df.set_index('Countries', inplace=True)
df
```

```python
df=df.replace('%','',regex=True).astype(float)
df
```

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = df.transpose()
df
```

```python
df=df.iloc[::-1]
df
```

```python
labels = ['Bofore 2000', '2000 to 2009', '2010 or Later']
height1=df['Europe']
height2=df['Asia']
height3=df['Africa']
height4=df['Oceania']
height5=df['Latin America']
height6=df['Northern America']
           
         

x = np.arange(len(labels))


fig, ax = plt.subplots(figsize=(20,8))
rects1 = ax.bar(x-0.25, height1, width=0.1, label='Europe',  color='#5cb85c')
rects2 = ax.bar(x-0.15, height2, width=0.1, label='Asia', color='#5bc0de')
rects3 = ax.bar(x-0.05, height3, width=0.1, label='Africa', color='#d9534f')
rects4 = ax.bar(x+0.05, height4, width=0.1, label='Oceania',  color='#d6832b')
rects5 = ax.bar(x+0.15, height5, width=0.1, label='Latin America', color='#6e48cf')
rects6 = ax.bar(x+0.25, height6, width=0.1, label='Northern America', color='#d62ba3')

ax.set_title("Percentage of Foreign-born Population in the U.S.", fontsize=20)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=15)
ax.legend(fontsize=11)
ax.set_ylabel('Percentage', fontsize=14) 

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 1), 
                    textcoords="offset points",
                    fontsize=14, 
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
autolabel(rects5)
autolabel(rects6)


plt.show()
```
